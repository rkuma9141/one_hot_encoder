# ONE HOT ENCODING

# One-Hot Encoding is a technique used to convert categorical data (labels/text) into numerical format so that machine learning algorithms can understand it.

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/cars (1).csv')
```

# one hot encoding using pandas

```python
pd.get_dummies(data=df, columns=['brand','fuel','owner'],dtype=int)
```

## one hot encodin using sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
df_trf=ohe.fit_transform(df[['brand','fuel','owner']])
```

```python
df_trf
```

```python
cols=ohe.get_feature_names_out()
```

```python
cols
```

```python
df_trf=pd.DataFrame(df_trf,columns=cols)
```
